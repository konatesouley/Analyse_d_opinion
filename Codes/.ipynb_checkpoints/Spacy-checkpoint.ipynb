```python
import spacy
from spacy import displacy


#nlp = spacy.load("fr_dep_news_trf")
nlp = spacy.load("fr_core_news_md")
```

```python
doc = nlp("Bonjour, je m'appelle Cynthia")
for token in doc:
    print("{0}\t{1}\t{2}\t{3}\t{4}\t{5}\t{6}\t{7}\t{8}".format(
        token.text,
        token.idx,
        token.lemma_,
        token.is_punct,
        token.is_space,
        token.shape_,
        token.pos_,
        token.tag_,
        token.ent_type_
    ))
```

```python
data = pd.read_excel('data_fichierSpacy.xlsx')
```

```python
data.head()
```

```python
spacy = pd.DataFrame(columns=['sujet','tokenpost','postfinal','sentiment','post','text','pos','tag'])
```

```python
ligne=0

for j in data['post']:
    donnee=data.iloc[ligne]
    doc = nlp(j)
    for token in doc:
        nv_ligne = pd.DataFrame([[donnee['sujet'], donnee['tokenpost'],donnee['postfinal'],donnee['sentiment'], donnee['post'], token.text,token.pos_,token.tag_]],
                      columns = spacy.columns)
        spacy = spacy.append(nv_ligne, ignore_index=True)    
    ligne=ligne+1
```

```python
spacy.head(5)
```

```python
spacy.to_excel('data_bogda_Spacy.xlsx',
           encoding = 'utf-8',
           header = True)
```
